```python
# Grouping by a categorical column 
grouped_product_category = sales_data.groupby("Product_Category")
print(grouped_product_category)
#Computing mean for the Quantity Sold column
mean_quantity_sold = grouped_product_category["Quantity_Sold"].mean()
print(mean_quantity_sold)
```

```python
# Importing the necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Line chart showing trends over time

# Convert the column to datetime
sales_data["Sale_Date"] = pd.to_datetime(sales_data["Sale_Date"])

# Sort the entire DataFrame by the 'Sale_Date' column
sales_data.sort_values("Sale_Date", inplace=True)

plt.figure(figsize=(10, 5))
plt.plot(sales_data["Sale_Date"], sales_data["Sales_Amount"], marker="o")
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
# Bar Chart showing the trends over time.
# To show the product category sold

# import matplotlib.pyplot as plt

# # Create a new column for total sales
# sales_data["Total_Sales"] = sales_data["Sales_Amount"] * sales_data["Quantity_Sold"]

# # Group by Product Category
# grouped_data = sales_data.groupby("Product_Category")["Total_Sales"].sum()

# # Plot the bar chart
# plt.bar(grouped_data.index, grouped_data.values, color="blue")
# plt.title("Total Sales by Product Category")
# plt.xlabel("Product Category")
# plt.ylabel("Total Amount Sold")
# plt.xticks(rotation=45)
# plt.tight_layout()
# plt.show()

#Create a new column for total sales
sales_data["Total_Sales"] = sales_data["Sales_Amount"] * sales_data["Quantity_Sold"]

#Grouped by Product Category
grouped_data = sales_data.groupby("Product_Category")["Total_Sales"].sum()

#Plot the bar chart
plt.bar(grouped_data.index, grouped_data.values, color='blue')
plt.title("Total Sales by Product Category")
plt.xlabel("Product_Category")
plt.ylabel("Total Amount Sold")
```

```python
plt.hist(sales_data["Total_Sales"], bins=10, color="skyblue", edgecolor="black")
plt.title("Distribution of Total Sales")
plt.xlabel("Total Sales")
plt.ylabel("Frequency")
```

```python
# Scatter plot
plt.scatter(sales_data["Sales_Amount"], sales_data["Quantity_Sold"], color="green", alpha=0.6)
plt.title("Sales Amount vs. Quantity Sold")
plt.xlabel("Sales Amount")
plt.ylabel("Quantity Sold")
plt.grid(True)
plt.tight_layout()
plt.show()
```
